Use the following code to generate data:

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
np.random.seed(1)
```

```python
x1 = np.random.normal(-4, 2, 1000)
x2 = np.random.normal(4, 2, 1000)
xs = np.append(x1, x2)
ys = np.asarray([0.] * len(x1) + [1.] * len(x2))
```

Use the dara `xs` and `ys` for training. The mode is $\hat{y} = \sigma (w*x+b)$ where $\sigma(\cdot)$ is the sigmoid function and `w` and `b` are the parameters to the model.

**Task**: Find the optimal parameters and plot the fitted model to the data.

**Hints**:
- You should use `tf.sigmoid()` function in the TensorFlow
- Use cross entropy loss: $\mathcal{L} = -y\log\hat{y}-(1-y)\log(1-\hat{y})$
- Use **gradient decent optimister** with learning rate parameter equal to 0.01
- Use the number of iterations equal to 1000

```python
ys_color = [['red' if y else 'green'] for y in ys]
# plot raw data xs
plt.scatter(xs, ys, c=np.squeeze(ys_color), alpha=0.2)
plt.show()
```

```python
# Contribute computational graph

# inputs
data_x = tf.placeholder(dtype=tf.float64)
data_y = tf.placeholder(dtype=tf.float64)

# training parameters
with tf.variable_scope('logistic', reuse=tf.AUTO_REUSE) as variable_scope:
    w = tf.get_variable(name="weights", shape=[], dtype=tf.float64, initializer=tf.random_normal_initializer())
    b = tf.get_variable(name="bias", shape=[], dtype=tf.float64, initializer=tf.random_normal_initializer())
    
# output
y_predict = tf.nn.sigmoid(tf.add(tf.multiply(w, data_x), b))

#loss function
cross_entropy = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=y_predict, labels=ys))
train_step = tf.train.GradientDescentOptimizer(learning_rate=0.01).minimize(cross_entropy)
```

```python
# training
iteration = 1000

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())  
    losses = []
    
    for i in range(iteration):
        _, loss = sess.run([train_step, cross_entropy], feed_dict={data_x: xs, data_y: ys})
        losses.append(loss)
    opt_w, opt_b = sess.run([w, b])
#     opt_w, opt_b = sess.run([w, b], feed_dict={data_x: xs, data_y: ys})
```

```python
plt.plot(np.arange(iteration), losses)
plt.show()
```

```python
from matplotlib.lines import Line2D

# Observation
plt.scatter(xs, ys, c="b")

# Model
xn = np.linspace(start=-10,stop=11,num=2000)
y_prob = np.exp(opt_w*xn+opt_b)/(1+np.exp(opt_w*xn+opt_b))
plt.plot(xn, y_prob,'r')

# Prediction
y_prob = np.exp(opt_w*xs+opt_b)/(1+np.exp(opt_w*xs+opt_b))
y_trained = np.where(y_prob > 0.5, 1, 0)

plt.scatter(xs[y_trained==1], y_trained[y_trained==1], c="g", alpha=0.2, label='class 1')
plt.scatter(xs[y_trained==0], y_trained[y_trained==0], c="y", alpha=0.2, label='class 0')

legend_elements = [Line2D([0], [0], marker='o', color='w', label='class 1',
                         markerfacecolor='g', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='class 0',
                          markerfacecolor='y', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='Observations',
                         markerfacecolor='b', markersize=10)]


plt.legend(handles=legend_elements)
plt.show()
```
